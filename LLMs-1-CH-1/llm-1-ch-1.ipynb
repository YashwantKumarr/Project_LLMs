```python
import torch
import torch.nn as nn

# ---------------------------------------------------------
# STEP 1: The Data
# ---------------------------------------------------------
# Input data (X) and the true answers (Y)
# The hidden rule we want the model to learn is simply: Y = X * 2
X = torch.tensor([[1.0], [2.0], [3.0], [4.0]])
Y = torch.tensor([[2.0], [4.0], [6.0], [8.0]])

# ---------------------------------------------------------
# STEP 2: The Model and The Robot Assistant
# ---------------------------------------------------------
# We create a simple model with 1 input and 1 output (just 1 weight and 1 bias)
model = nn.Linear(in_features=1, out_features=1)

# We hire our Robot Assistant (Optimizer) and give it the keys to the model's dials
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

# We define how we will measure our Error (Mean Squared Error)
loss_function = nn.MSELoss()

# ---------------------------------------------------------
# STEP 3: The Training Loop (Practicing 50 times)
# ---------------------------------------------------------
epochs = 500

for epoch in range(epochs):
    
    # Action A: Make a prediction (Play the soundboard)
    predictions = model(X)
    
    # Action B: Calculate the Error (How bad does the music sound?)
    loss = loss_function(predictions, Y)
    
    # Action C: Throw away old sticky notes from the previous loop!
    optimizer.zero_grad()
    
    # Action D: Do calculus and print new sticky notes (Gradients)
    loss.backward()
    
    # Action E: The Robot Assistant reads the notes and updates the dials!
    optimizer.step()
    
    # Print the progress every 10 steps so we can watch it learn
    if (epoch + 1) % 10 == 0:
        print(f"Practice Round {epoch + 1} | Error (Loss): {loss.item():.4f}")

# ---------------------------------------------------------
# STEP 4: The Final Test
# ---------------------------------------------------------
print("\n--- Training Complete! ---")

# Let's ask it to predict a number it has NEVER seen before: 10.0
test_number = torch.tensor([[10.0]])

# Turn off the calculus engine since we are just testing, not training
with torch.no_grad():
    final_guess = model(test_number)

print(f"If the input is 10.0, the model guesses: {final_guess.item():.4f}")
print("(The perfect answer would be 20.0)")
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

# ---------------------------------------------------------
# STEP 1: The Data
# X = [x_coordinate, y_coordinate]
# Y = Category (0 is Red, 1 is Blue)
# ---------------------------------------------------------
X_train = torch.tensor([
    [-1.2, 3.1], [-0.9, 2.9], [-0.5, 2.6], # These are Red dots (0)
    [ 2.3,-1.1], [ 2.7,-1.5]               # These are Blue dots (1)
])
y_train = torch.tensor([0, 0, 0, 1, 1])

# ---------------------------------------------------------
# STEP 2: The Model and Robot Assistant
# ---------------------------------------------------------
# 2 inputs (x, y coordinates) -> 2 outputs (Red score, Blue score)
model = nn.Linear(in_features=2, out_features=2) 

optimizer = torch.optim.SGD(model.parameters(), lr=0.5)

# ---------------------------------------------------------
# STEP 3: The 5-Step Training Loop
# ---------------------------------------------------------
epochs = 3

for epoch in range(epochs):
    
    # 1. Predict (Outputs 2 scores for every dot)
    logits = model(X_train)
    
    # 2. Calculate Error using Cross Entropy
    loss = F.cross_entropy(logits, y_train)
    
    # 3. Trash old sticky notes
    optimizer.zero_grad()
    
    # 4. Print new sticky notes
    loss.backward()
    
    # 5. Robot twists the dials
    optimizer.step()
    
    print(f"Epoch: {epoch+1:03d}/{epochs:03d} | Error: {loss.item():.2f}")

# ---------------------------------------------------------
# STEP 4: Evaluating the Model (Did it actually learn?)
# ---------------------------------------------------------
print("\n--- Let's see its final predictions! ---")

with torch.no_grad():
    raw_scores = model(X_train)

# Softmax turns the raw scores into percentages (0% to 100%)
probabilities = torch.softmax(raw_scores, dim=1)
print("Probabilities (Red% vs Blue%):\n", probabilities)

# Argmax picks the category with the highest percentage
predictions = torch.argmax(probabilities, dim=1)
print("\nModel's Final Guesses:", predictions)
print("The True Answers were:  ", y_train)

# Calculate Accuracy
correct_guesses = torch.sum(predictions == y_train)
print(f"\nAccuracy: {correct_guesses.item()} / 5 correct!")
```
